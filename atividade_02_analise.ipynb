##Questão 2

Para calcular a média de um Array Numpy se usa a função average(). Sabendo disso, imprimir a média de Attack de todos os pokemons

Documentação:
- https://numpy.org/doc/stable/reference/generated/numpy.average.html

```python
array_pokemon2 = np.loadtxt('Pokemon.csv', delimiter=',', skiprows=1, usecols=6)

media_attack = np.average(array_pokemon2)

print("Média dos valores da coluna 'Attack':", media_attack)


```

### Questão 3:
Sabendo que para encontrar o maior e menor valor de um Array Numpy se usa respectivamente as funções max() e min(). Encontre o maior e menor valor de Hit Points (HP) dos pokemons.

Documentação:
- https://numpy.org/doc/stable/reference/generated/numpy.max.html
- https://numpy.org/doc/stable/reference/generated/numpy.min.html

### Questão 4:
Sabendo que o desvio padrão de um array numpy é calculado através da função std(), encontre o desvio padrão do Attack dos pokemons.

Documentação:
- https://numpy.org/doc/stable/reference/generated/numpy.std.html

```python
array_pokemon4 = np.loadtxt('Pokemon.csv', delimiter=',', skiprows=1, usecols=6)

desvio_padrao = np.std(array_pokemon4)

print("Desvio padrão do Attack:", desvio_padrao)
```

### Questão 5:
Imprima o nome de todos os pokemons que possuem Attack acima da média.

```python

array_nome_pokemon = np.loadtxt('Pokemon.csv', delimiter=',', skiprows=1, usecols=1, dtype='str')

array_attack = np.loadtxt('Pokemon.csv', delimiter=',', skiprows=1, usecols=6)


media_attack_pokemon = np.mean(array_attack)

acima_da_media = np.where(array_attack > media_attack_pokemon)

nomes_acima_da_media = array_nome_pokemon[acima_da_media]


# Imprime a média calculada
print("Média de Attack:", media_attack_pokemon)
print("Pokémons com Attack acima da média:")
print(nomes_acima_da_media)


```

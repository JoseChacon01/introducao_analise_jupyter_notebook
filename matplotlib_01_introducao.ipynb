## plot()

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
x = [1,2,3,4,5]
y = [-2,5,200,50,-24]

plt.plot(x,y, label="Notas")
plt.legend()
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.xticks([1,2,3])
plt.yticks([-30,0,30,60,90,120,150,180,210])
plt.show()
```

### Propriedades de Linha

https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html

```python
x = [1,2,3,4,5]
y = [-2,5,200,50,-24]

plt.plot(x,y, label="Notas", linestyle="--", color="m", linewidth="3")
plt.legend()
plt.title("Gráfico da aula")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.xticks([1,2,3])
plt.yticks([-30,0,30,60,90,120,150,180,210])
plt.show()
```

## Tipos de Gráficos

### scatter - gráfico de dispersão

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html

```python
plt.scatter(x,y)
plt.show()
```

### bar - gráfico de barras

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html

```python
plt.bar(x,y)
plt.show()
```

### pie - gráfico de pizza

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html

```python
z = [25,10,8,78]
plt.pie(z)
plt.show()
```

## Mudando o tipo de marcador do gráfico

https://matplotlib.org/stable/api/markers_api.html

```python
plt.scatter(x,y, marker="v")
plt.show()
```

## Mudando os limites do gráfico

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.axis.html

```python
plt.plot(x,y)
plt.axis(xmin=-1, xmax=8, ymin=-50, ymax=300)
plt.show()
```

## Na prática - dataset Pokemon e UFO

```python
import pandas as pd
pokemon = pd.read_csv('Pokemon.csv')
pokemon
```

```python
pokemon_attack = pokemon[['Name', 'Attack']]
pokemon_attack
```

```python
x = pokemon_attack['Name']
y = pokemon_attack['Attack']
plt.bar(x,y)
plt.show()
```

```python
pokemons_maiores = pokemon_attack.sort_values('Attack', ascending=False).head(5)

x = pokemons_maiores ['Name']
y = pokemons_maiores ['Attack']
plt.bar(x,y)
plt.show()
```

```python
pokemons_maiores = pokemon_attack.sort_values('Attack', ascending=False).head(5)

x = pokemons_maiores ['Name']
y = pokemons_maiores ['Attack']
plt.figure(figsize=(15,5))
plt.bar(x,y)
plt.show()
```

```python
pokemons_maiores = pokemon_attack.sort_values('Attack', ascending=False).head(5)

x = pokemons_maiores ['Name']
y = pokemons_maiores ['Attack']
plt.bar(x,y)
plt.xticks(rotation=50, ha='right')
plt.show()
```

```python
y_attack = pokemon['Attack'].head(5)
y_defense = pokemon['Defense'].head(5)
x = pokemon['Name'].head(5)

plt.plot(x, y_attack, label='Attack')
plt.plot(x, y_defense, label='Defense')
plt.legend()
plt.xticks(rotation=50, ha='right')
plt.show()
```
